```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('8-all-at-once.txt')

# 10 colors in cycle
#plt.figure(figsize=(8,10))

plt.plot(df.fitMax)
        
plt.legend(loc=2, bbox_to_anchor=(1.01, 1.0))
plt.ylim(0,1)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('TPG Max Fitness Scores per Generation')
plt.show()
```

```python
df = pd.read_csv('8-all-at-once-window-2.txt')

envDfs = []
for envs in pd.unique(df.envs):
    envDfs.append(df.loc[df['envs'] == envs])

# 10 colors in cycle
plt.figure(figsize=(8,10))
cnt = 0
lineStyles = ['-', '--', '-.']
lsi = 0
for edf in envDfs:
    cnt += 1
    plt.plot(edf.fitMax, label=edf.envs.iloc[0]) 
             #label=edf.envs.iloc[0], ls=lineStyles[lsi])
    if cnt == 10:
        cnt = 0
        lsi += 1
        
plt.legend(loc=2, bbox_to_anchor=(1.01, 1.0))
plt.ylim(0,1)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('TPG Max Fitness Scores per Generation')
plt.show()
```

```python
df = pd.read_csv('8-all-at-once-window-2.txt')

envsFits = {}

for i in range(len(df)):
    row = df.iloc[i]
    curEnvs = row['envs'].split('/')
    if len(curEnvs) < 2:
        continue
        
    for env in curEnvs:
        if env not in envsFits:
            envsFits[env] = []
        envsFits[env].append(row['fitMax'])
        
plt.boxplot([envsFits[env] for env in sorted(envsFits.keys())], labels=sorted(envsFits.keys()))
plt.xticks(rotation='vertical')
plt.ylim(0,1)
plt.title('Scores from Environment Groups')
plt.xlabel('Environment')
plt.ylabel('Score (Normalized per Generation)')
plt.show() 
```
