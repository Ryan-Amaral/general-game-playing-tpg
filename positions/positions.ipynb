```python
import pandas as pd
```

```python
df = pd.read_csv('30-noshrink.txt')

mDict = {}
for thing in df.iloc[0,3:].tolist():
    mDict[thing] = {}

for thing in df.iloc[0,3:33].tolist():
    try:
        mDict[thing]['1st'] += 1
    except:
        mDict[thing]['1st'] = 1
        
for thing in df.iloc[0,33:63].tolist():
    try:
        mDict[thing]['2nd'] += 1
    except:
        mDict[thing]['2nd'] = 1
        
for thing in df.iloc[0,63:93].tolist():
    try:
        mDict[thing]['3rd'] += 1
    except:
        mDict[thing]['3rd'] = 1
        
for thing in df.iloc[0,93:123].tolist():
    try:
        mDict[thing]['4th'] += 1
    except:
        mDict[thing]['4th'] = 1
        
for thing in df.iloc[0,123:153].tolist():
    try:
        mDict[thing]['5th'] += 1
    except:
        mDict[thing]['5th'] = 1
        
for key in mDict:
    print(str(key) + ': ' + str(mDict[key]))
```
