```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
df = pd.read_csv('8-all-at-once-window-4-2.txt')

def isEmptyRow(row):
    for i in range(3, len(row)):
        if row[i] != -999999:
            return False
    
    return True

envCounts = {'Alien-v0': [0]*len(pd.unique(df['tpgGen'])),
            'Asteroids-v0': [0]*len(pd.unique(df['tpgGen'])),
            'Atlantis-v0': [0]*len(pd.unique(df['tpgGen'])),
            'BankHeist-v0': [0]*len(pd.unique(df['tpgGen'])),
            'BattleZone-v0': [0]*len(pd.unique(df['tpgGen'])),
            'Bowling-v0': [0]*len(pd.unique(df['tpgGen'])),
            'Boxing-v0': [0]*len(pd.unique(df['tpgGen'])),
            'Centipede-v0': [0]*len(pd.unique(df['tpgGen']))}

for row in [df.iloc[i] for i in range(len(df)) if not isEmptyRow(df.iloc[i])]:
    for env in envCounts.keys():
        if row['score' + env] != -999999:
            envCounts[env][int(row['tpgGen']/10)-1] += 1

plt.figure(figsize=(12,6))
            
for env in sorted(envCounts.keys()):
    plt.plot(envCounts[env], label=env)
    
plt.legend(loc=2, bbox_to_anchor=(1.01, 1.0))
#plt.ylim(0,1)
plt.xlabel('Generation')
plt.ylabel('Number of Agents')
plt.title('Amount of Agents with Scores for Environment')
plt.show()
```

```python
df = pd.read_csv('8-all-at-once-window-4-2.txt')

gamesPerAgent = {
            1: [0]*len(pd.unique(df['tpgGen'])),
            2: [0]*len(pd.unique(df['tpgGen'])),
            3: [0]*len(pd.unique(df['tpgGen'])),
            4: [0]*len(pd.unique(df['tpgGen'])),
            5: [0]*len(pd.unique(df['tpgGen'])),
            6: [0]*len(pd.unique(df['tpgGen'])),
            7: [0]*len(pd.unique(df['tpgGen'])),
            8: [0]*len(pd.unique(df['tpgGen']))}

envs = ['Alien-v0','Asteroids-v0','Atlantis-v0','BankHeist-v0',
            'BattleZone-v0','Bowling-v0','Boxing-v0','Centipede-v0']

for row in [df.iloc[i] for i in range(len(df)) if not isEmptyRow(df.iloc[i])]:
    gamesPlayed = 0
    for env in envs:
        if row['score' + env] != -999999:
            gamesPlayed += 1
            
    gamesPerAgent[gamesPlayed][int(row['tpgGen']/10)-1] += 1

plt.figure(figsize=(8,4))
    
for i in range(1,9):
    plt.plot(gamesPerAgent[i], label=str(i) + ' Games Played')
    
plt.legend(loc=2, bbox_to_anchor=(1.01, 1.0))
#plt.ylim(0,1)
plt.xlabel('Generation')
plt.ylabel('Number of Agents')
plt.title('Amount of Agents with Scores on Various Games')
plt.show()
```

```python
df = pd.read_csv('8-all-at-once-window-4-2.txt')

envs = ['Alien-v0','Asteroids-v0','Atlantis-v0','BankHeist-v0',
            'BattleZone-v0','Bowling-v0','Boxing-v0','Centipede-v0']
envPairs = list(combinations(envs, 2))
#print(len(list(combinations(envs,4))))

gens = 5

envPoints = {}
for env in envs:
    envPoints[env] = [[]]*gens

for row in [df.iloc[i] for i in range(len(df)) if not isEmptyRow(df.iloc[i]) and df.iloc[i]['tpgGen']%50 == 0]:
    genNum = int(row['tpgGen'] / 50)-1
    for env in envs:
        envPoints[env][genNum].append((row['score'+env] - soloRandomScores[env])/
                                (soloTpgScores[env] - soloRandomScores[env]))
```

```python
colors = ['C0', 'C1', 'C2', 'C3', 'C4']

#plt.figure(figsize=(10,10))


for i, envPair in enumerate(envPairs):
    plt.ylim(0,1)
    plt.xlim(0,1)
    plt.xlabel(envPair[0])
    plt.ylabel(envPair[1])
    for gen in range(gens):
        plt.scatter(envPoints[envPair[0]][gen], envPoints[envPair[1]][gen], c=colors[gen], s=20-(gen*5))
        
    plt.show()
```
